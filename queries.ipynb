```python
#query1: 

query1 = '''select r.app_name, count(distinct user_reviews) as max_unique_reviews, avg(polarity) as avg_polarity, avg(subjectivity) as avg_subjectivity
from reviews r join sentiments s
on r.app_name = s.app_name
group by r.app_name
having count(distinct user_reviews) = (
select max(mycount) from(select app_name, count(distinct user_reviews) mycount
from reviews
group by app_name)abc)'''
#Creating a cursor object using the cursor() method
cursor = conn.cursor()

#Executing an MYSQL function using the execute() method
cursor.execute(query1)

# Fetch a single row using fetchone() method.
query1_result = cursor.fetchall()

#display one row in one line
for x in query1_result:
   print(x)
```

```python
#query2: 

query2 = '''	select user_reviews
from reviews 
where app_name in (
select app.app_name from sentiments s
join applications app on s.app_name = app.app_name
where sentiment like '%Positive'
group by 1
having (count(s.sentiment))
in
(
select max(cnt1) from
(
select app.app_name as cat1, count(s.sentiment) as cnt1
from applications app
join sentiments s
on app.app_name = s.app_name
	where sentiment like '%Positive'
group by app.app_name
)abs
)
)
LIMIT 5
'''
#Creating a cursor object using the cursor() method
cursor = conn.cursor()

#Executing an MYSQL function using the execute() method
cursor.execute(query2)

# Fetch a single row using fetchone() method.
query2_result = cursor.fetchall()

#display one row in one line
for x in query2_result:
   print(x)
```

```python
#query5: 

query5 = '''select app.app_name, req.app_size
from applications app left join requirements req
on app.app_id = req.app_id
where app.app_name in(
select cat1 from
(
select app.app_name as cat1, count(s.sentiment) as cnt1
from applications app
join sentiments s
on app.app_name = s.app_name
	where sentiment like '%Positive'
group by app.app_name
)abs
where cnt1 >= 20)
order by 2 
LIMIT 5
'''

cursor.execute(query5)

# Fetch a single row using fetchone() method.
query5_result = cursor.fetchall()

#display one row in one line
for x in query5_result:
   print(x)
```
